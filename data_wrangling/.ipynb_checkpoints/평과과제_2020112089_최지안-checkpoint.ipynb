# 이상치 확인 및 제거

```python
# numerical feature들에 대해 이상치 확인 ('Height, Weight, Temperature, Heart Rate, Blood Pressure, Cholesterol)

numerical_features = ['Height', 'Weight', 'Temperature', 'Heart Rate', 'Blood Pressure', 'Cholesterol']

plt.figure(figsize=(20, 15))

for i, col in enumerate(numerical_features):
    plt.subplot(3,2,i+1)
    sns.boxplot(x=df[col], color = 'darkviolet')

plt.show()
```

```python
# IQR을 사용하여 이상치 제거
for col in numerical_features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
# numerical feature들에 대해 이상치 확인 ('Height, Weight, Temperature, Heart Rate, Blood Pressure, Cholesterol)

numerical_features = ['Height', 'Weight', 'Temperature', 'Heart Rate', 'Blood Pressure', 'Cholesterol']

plt.figure(figsize=(20, 15))

for i, col in enumerate(numerical_features):
    plt.subplot(3,2,i+1)
    sns.boxplot(x=df[col], color = 'darkviolet')

plt.show()
```

# 결측치 확인 및 제거

```python
len(df)
```

```python
df.isna().sum()
```

```python
# 성별과 관련없는 Numerical 결측치들은 중앙값으로 대체
numerical_features = ['Age','Temperature', 'Heart Rate', 'Blood Pressure', 'Cholesterol']

for col in numerical_features:
    df[col] =  df[col].fillna(df[col].median())
    print(f'median of {col} = ',df[col].median())
```

```python
df[numerical_features]
```

### 성별과 관련있는 Numerical 결측치들은 각 성별의 중앙값으로 대체

### categorical 결측치들은 다른 변수들과의 상관관계를 비교 후 대체

```python
# corr를 계산하기 위해 'Diabets'와 'Smoking'을 각각 0,1로 원핫 인코딩
def binarize(value):
    if value == 'Yes':
        return 1
    elif value == 'Female':
        return 1
    else:
        return 0
    
df['Diabetes'] = df['Diabetes'].apply(binarize)
df['Smoking'] = df['Smoking'].apply(binarize)
df['Gender'] = df['Gender'].apply(binarize)
df[['Gender','Diabetes', 'Smoking']]
```
