# 이상치 확인 및 제거

```python
# numerical feature들에 대해 이상치 확인 ('Height, Weight, Temperature, Heart Rate, Blood Pressure, Cholesterol)

numerical_features = ['Height', 'Weight', 'Temperature', 'Heart Rate', 'Blood Pressure', 'Cholesterol']

plt.figure(figsize=(15, 15))

for i, col in enumerate(numerical_features):
    plt.subplot(3,2,i+1)
    sns.boxplot(x=df[col], color = 'darkviolet')

plt.show()
```

```python
# IQR을 사용하여 이상치 제거
for col in numerical_features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
# numerical feature들에 대해 이상치 확인 ('Height, Weight, Temperature, Heart Rate, Blood Pressure, Cholesterol)

numerical_features = ['Height', 'Weight', 'Temperature', 'Heart Rate', 'Blood Pressure', 'Cholesterol']

plt.figure(figsize=(20, 15))

for i, col in enumerate(numerical_features):
    plt.subplot(3,2,i+1)
    sns.boxplot(x=df[col], color = 'darkviolet')

plt.show()
```

```python
print("이상치 제거 후 길이 : ", len(df))
```

# 결측치 확인 및 대체

```python
df.isna().sum()
```

## 결측치들을 채워주기 전에 각 변수들 간의 상관관계 분석

```python
# corr를 계산하기 위해 Diabetes','Smoking'을 각각 0,1로 binarization
def binarize(value):
    if value == 'Yes':
        return 1
    else:
        return 0
    
df['Diabetes'] = df['Diabetes'].apply(binarize)
df['Smoking'] = df['Smoking'].apply(binarize)
df[['Diabetes', 'Smoking']]
```

```python
plt.figure(figsize=(20,20))
# Student ID를 제외한 feature들끼리 상관관계 분석
sns.heatmap(df.corr(), annot=True, fmt=".5f", cmap='Blues')
plt.show()
```

```python
# 각 변수들 간의 상관관계가 유의미하지 않으므로 Numerical 결측치들은 각 feature의 중앙값으로 대체
numerical_features = ['Age','Temperature', 'Heart Rate', 'Blood Pressure', 'Cholesterol']

for col in numerical_features:
    df[col] =  df[col].fillna(df[col].median())
    print(f'median of {col} = ',df[col].median())
```

## 수치상으로 대체하기 어려운 feature가 포함된 data들은 제거

```python
cannot_replaceable_features = ['Age', 'Gender' ,'Blood Type', 'Diabetes', 'Smoking']
```

```python
# 결측치가 있는 행 삭제
for feature in cannot_replaceable_features:
    df = df.dropna(subset=[feature])
```

```python
len(df)
```

## 'Student ID'와 'BMI' 재할당

```python
df.isnull().sum()
```

```python
# Height와 Weight 기반으로 'BMI' 재계산
df['BMI'] = (df['Weight'] / pow(df['Height'], 2)) * 10000
```

```python
# index 기반으로 'Student ID' 재할당
df['Student ID'] = np.arange(1, len(df) + 1, dtype='float64')
```

```python
# 결측치 유무 확인
df.isnull().sum()
```

```python
# 중복값 유무 확인
df[df.duplicated()]
```

```python
# binarization 되었던 'Diabetes','Smoking'을 각각 'Yes', 'No'로 재할당
def binarize(value):
    if value == 1:
        return 'Yes'
    else:
        return 'No'
    
df['Diabetes'] = df['Diabetes'].apply(binarize)
df['Smoking'] = df['Smoking'].apply(binarize)
df[['Diabetes', 'Smoking']]
```
